Task 5 – Variable Scope

```python
x = 10   # global variable

def check_scope():
    print(f"Inside function, x = {x}")
    y = 20   # local variable
    print(f"Inside function, y = {y}")

check_scope()
print(y)
```

Task 6 – Variable Length Arguments (*args)

```python
def total_numbers(*numbers):
    total = sum(numbers)
    return total

print(f"Total = {total_numbers(2,3,4,5)}")
```

Task 7 – Keyword Variable Arguments (**kwargs)

```python
def student_info(**data):
    for key, value in data.items():
        print(f"{key}: {value}")

student_info(name="Sundas", age=20, course="AI")
```

Task 9 – Lambda with List

```python
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x*x, numbers))
print(f"Squares List = {squares}")
```

Task Mini Project: Student Marks Analyzer

```python
def calculate_total(marks):
    return sum(marks)

def calculate_average(marks):
    return sum(marks) / len(marks)

def highest_mark(marks):
    return max(marks)

# Input marks
marks = [int(input("Enter marks: ")) for _ in range(5)]

total = calculate_total(marks)
average = calculate_average(marks)
highest = highest_mark(marks)

print("\n----- Result -----")
print(f"Total Marks = {total}")
print(f"Average Marks = {average}")
print(f"Highest Marks = {highest}")
```

Mini-Project: &quot;AI Number Guessing Game Lite&quot;

```python
import random

secret = random.randint(1, 50)
attempts = 7
count = 0

print("\n AI Number Guessing Game")

while count < attempts:
    guess = int(input("Enter your guess (1-50): "))
    count += 1

    if guess < 1 or guess > 50:
        print("Please enter number between 1 and 50")
        continue

    if guess == secret:
        print(f"You win in {count} attempts! AI training level: Beginner to Intermediate")
        break
    elif guess > secret:
        print("Too high!")
    else:
        print("Too low!")

    print(f"Remaining attempts: {attempts - count}")

else:
    print(f"\nYou lost! The number was {secret}")

print("--")
```
