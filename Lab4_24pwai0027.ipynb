AI Number Guessing Game Lite

```python
import random

# Generating random numbers between 1 and 50
secret_number = random.randint(1, 50)

attempts = 7

print("AI Number Guessing Game Lite")
print("Guess the number between 1 and 50")

for i in range(1, attempts + 1):

    try:
        # Taking user input
        guess = int(input(f"\nAttempt {i}: Enter your guess: "))

        # Checking range
        if guess < 1 or guess > 50:
            print(f"Please enter a number between 1 and 50")
            continue

        # Correct guess
        if guess == secret_number:
            print(f"You win in {i} attempts!")
            print(f"AI training level: Beginner → Intermediate")
            break

        # Guess too high
        elif guess > secret_number:
            print(f"Too high!")

        # Guess too low
        else:
            print(f"Too low!")

        # Remaining attempts
        remaining = attempts - i
        print(f"Remaining attempts: {remaining}")

    except ValueError:
        print(f"Invalid input! Please enter an integer.")

# If player loses all attempts
else:
    print(f"\nGame Over!")
    print(f"The correct number was: {secret_number}")
```
