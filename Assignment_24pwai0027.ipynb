NumPy for AI - Mini Project Assignment

```python
import numpy as np
```

```python
# Dataset of 10 students
# Math, Science, English, Computer, Result

data = np.array([
    [78, 85, 80, 90, 1],
    [45, 50, 40, 55, 0],
    [88, 92, 85, 95, 1],
    [60, 70, 65, 75, 1],
    [30, 35, 25, 40, 0],
    [95, 90, 93, 97, 1],
    [55, 58, 60, 62, 1],
    [40, 45, 42, 38, 0],
    [82, 78, 88, 84, 1],
    [48, 52, 50, 46, 0]
])

print("Dataset:\n")
print(data)
```

```python
# Features
X = data[:, 0:4]

# Labels
y = data[:, 4]

print("Features (X):\n")
print(X)

print("\nLabels (y):\n")
print(y)
```

```python
average_marks = np.mean(X)
maximum_marks = np.max(X)
minimum_marks = np.min(X)

print("Average Marks:", average_marks)
print("Maximum Marks:", maximum_marks)
print("Minimum Marks:", minimum_marks)
```

```python
passed_students = np.sum(y == 1)
failed_students = np.sum(y == 0)

print("Passed Students:", passed_students)
print("Failed Students:", failed_students)
```

```python
above_80 = X[X > 80]
below_50 = X[X < 50]

print("Marks Above 80:")
print(above_80)

print("\nMarks Below 50:")
print(below_50)
```

```python
normalized_data = X / 100

print("Normalized Dataset:\n")
print(normalized_data)
```

```python
# average marks of each student
student_average = np.mean(X, axis=1)

# If average >= 50 → Pass (1)
# Else → Fail (0)

predictions = np.where(student_average >= 50, 1, 0)

print("Predictions:\n")
print(predictions)
```

```python
accuracy = np.mean(predictions == y) * 100

print("Prediction Accuracy:", accuracy, "%")
```
