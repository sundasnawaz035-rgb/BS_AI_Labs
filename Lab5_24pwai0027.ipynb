Password Strength Checker

```python
attempts = 5

while attempts > 0:
    password = input(f"Enter your password ({attempts} attempts left): ")

    length = len(password) >= 8
    digit = any(char.isdigit() for char in password)
    uppercase = any(char.isupper() for char in password)

    score = length + digit + uppercase

    if score == 0:
        print(f"Password Strength: Very Weak")
    elif score == 1:
        print(f"Password Strength: Weak")
    elif score == 2:
        print(f"Password Strength: Moderate")
    else:
        print(f"Password Strength: Strong")
        print(f"You created a strong password!")
        print(f"AI Training Level: Beginner → Intermediate")
        break

    attempts -= 1

if attempts == 0:
    print(f"Final Result: Password is not strong enough.")
```
